## Filtrando dados

```python
df.loc[(df['Type 1'] == 'Grass') & (df['HP'] > 78)] 
#lembrar de colocar em parenteses aqui ele mostra os grama com vida acima de 78
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['HP'] > 78)] 
#crio um df novo apenas com o filtro
new_df = new_df.reset_index(drop = True)
#desta forma o novo df cria uma nova coluna de index o parametro apaga o index antigo
# da pra fazer sem atribuir o df na forma de new_df.reset_index(drop = True, inplace = True)
new_df
```

```python
df.loc[df['Name'].str.contains('Mega')] 
#aqui eu busco tudo que no nome tem mega, eu posso negar isso com ~ 
```

```python
import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)] 
#aqui ele procura linhas que o tipo 1 seja fire ou grass mas perceba que usamos dentro da string, isso graças a essa biblioteca.
# a expressão regex=True habilita procurar dessa forma
# a expressão flags=re.I faz a procura ignorar letras maiúsculas.
# o sinal | representa ou.
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)] 
# aqui ele procura nomes que começam com pi
# o ^ informa que a procura tem que começar com as letras especificadas
# o [a-z] informa que pode ser qualquer letra depois de pi
# * informa que pode incluir nada
```
